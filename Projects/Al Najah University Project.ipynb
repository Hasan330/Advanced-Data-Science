```python
#Create a function that takes in excel_file and sheet_name and returns the extracted dataframe
#Example usage df = Student_Extractor(xlsfile, 1)

def Student_Extractor(excel_file, index):
    #  Parse the file
    names = excel_file.sheet_names
    #print("All Sheet Names:", names)
    print("Parsing Sheet with name:", names[index-1])
    df = excel_file.parse(sheetname= names[index-1])
    
    #  Drop first 2 columns and 10 rows
    df_temp = df
    df_temp.drop(df_temp.columns[[0]], axis=1, inplace=True)
    df_temp.drop([0,1,2,3,4,5,6,7], axis=0, inplace=True)
    print("Dataframe" , index , "cleaned !")
    
    #  Rename the columns
    column_name = ['Student_Number', 'Student_Gender', 'First_Exam', 'Second_Exam', 'Final_1', 'Final_2', 'Total_Grade', 'Teacher_Name', 'Teacher_Number', 'Subject', 'Subject_Id', 'Section', 'Year', 'Semester']
    df_temp.columns = column_name
    
    #  Return the Data Frame
    return df_temp
```

## Clean Data 

```python
def replace999_and_clean_total(dataframe): 
    dataframe = dataframe.replace(9999, 0)                                      #Didn't Attend
    dataframe['Total_Grade'] = dataframe['Total_Grade'].replace(4, 30)          #Withdrew
    dataframe['Total_Grade'] = dataframe['Total_Grade'].replace(8, 30)          #Withdrew

    return dataframe
```

## Normalize Data

```python
def normalize_grades(df, full1, full2, full3, full4, weight1, weight2, weight3, weight4, project = 0, projectWeight=0):
    #First_Exam / full1 * weight1
    df['First_Exam']   = df['First_Exam']  / full1 * weight1
    df['Second_Exam']  = df['Second_Exam'] / full2 * weight2
    df['Final_1']      = df['Final_1']     / full3 * weight3
    df['Final_2']      = df['Final_2']     / full4 * weight4
    return df
```

# Machine Learning

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

## Multivariate Random Forest Regressor

```python
#Selecting numeric values from the dataframe
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numeric_df = df_attempted.select_dtypes(include=numerics)
```

```python
X_multi = numeric_df.loc[:, ['Student_Gender', 'First_Exam', 'Second_Exam', 'Teacher_Number', 'Year', 'Project']]
print(X_multi[:5])
```

### Predicting Students' Grade in the Subject

```python
y= numeric_df.Total_Grade

# Model
rfr = RandomForestRegressor(100)

## Fit Model (Format array for y with [:,None])
rfr.fit(X_multi, y)

## Set predicted points
predicted_final_grades = rfr.predict(X_multi)

print(predicted_final_grades[:5])

##Check Model score (accuracy/error)
sc= rfr.score(X_multi, y, sample_weight=None)
print("\nAccuracy score for predicting the student's total grade=", sc*100, "%")
```

### Predicting Students' Grade in the Final Exam

```python
#Changing y to predict the mark the student will get in the Final Exam
y= numeric_df.Final

# Model
rfr = RandomForestRegressor(100)

## Fit Model (Format array for y with [:,None])
rfr.fit(X_multi, y)

## Set predicted points
predicted_final_grades = rfr.predict(X_multi)

print(predicted_final_grades[:5])

##Check Model score (accuracy/error)
sc= rfr.score(X_multi, y, sample_weight=None)
print("\nAccuracy score for predicting the student's mark in Final Exam=", sc*100, "%")
```
